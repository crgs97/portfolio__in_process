```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc_data = sc.fit_transform(data)
data1 = pd.DataFrame(sc_data, columns = data.columns)
```

                                                                                                        Model Development.

```python
# Initializing the necessary Libraries for Model Deploying.

import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

from sklearn.cluster import KMeans

from sklearn.cluster import DBSCAN

from sklearn.metrics import silhouette_score
```

    -> AGGLOMERATIVE HIERARCHIAL CLUSTERING MODEL:

```python
plt.figure(figsize=(24,10), dpi=200, facecolor ='darkgrey')
dendrogram = sch.dendrogram(sch.linkage(data1, method='ward'))
plt.title('Dendrogram')
plt.ylabel('Euclidean Distance')
plt.xlabel('DATAPOINTS')
plt.grid()
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward') #with reference to dendrogram plotted we have got the number of clusters as "3".
y_hc = hc.fit_predict(data1)
hc1 = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
y_hc1 = hc1.fit_predict(data1)

sil_avg = silhouette_score(data1, y_hc)
sil_avg1 = silhouette_score(data1, y_hc1)

aghc_labels = hc.labels_
```

```python
plt.scatter(data1['Alcohol'], data1['Color_Intensity'], c = aghc_labels, cmap = 'Accent', edgecolors = 'b')
plt.scatter(data1['Alcohol'], data1['Magnesium'], c = aghc_labels, cmap = 'Accent', edgecolors = 'b')
plt.title('AGGLOMERATIVE HIERARCHIAL CLUSTERING')
plt.xlabel('Cluster Density Range')
plt.show()
```

    -> K-MEANS MODEL:


    => ELBOW METHOD:- {For finding out Which is the Optimum Number of Clusters}

```python
wcss = [] 

for i in range(1, 13): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(data1) 
    wcss.append(kmeans.inertia_)
    centroids=kmeans.cluster_centers_
    labels=kmeans.labels_
plt.figure(figsize = (24,6))
plt.plot(wcss, '--bo', markersize = 7)
plt.title('The Elbow Method (Using WCSS).')
plt.xlabel('Cluster or K Values.')
plt.ylabel('WCSS Values')
plt.show()
```

```python
kmeans = KMeans(algorithm = 'auto', n_clusters=2)
kmeans.fit(data1)
km_labels = kmeans.labels_
y_km = kmeans.fit_predict(data1)
sil_kmeans = silhouette_score(data1, y_km)
```

```python
clusters = kmeans.predict(data1)
clusters
```

```python
plt.scatter(data1['Alcohol'], data1['Color_Intensity'], c = km_labels, cmap = 'Accent', edgecolors = 'b')
plt.scatter(data1['Alcohol'], data1['Magnesium'], c = km_labels, cmap = 'Accent', edgecolors = 'b')
plt.title('KMeans Clustering')
plt.xlabel('Cluster Density Range')
plt.show()
```

    -> Density-Based Spatial Clustering of Applications with Noise (DBSCAN):

```python
d = data1.iloc[:,[0,12]].values
```

```python
dbscan = DBSCAN(algorithm = 'auto', eps=0.5, metric = 'euclidean', leaf_size = 30, min_samples = 10)
dbscan.fit(d)
dbscan_labels = dbscan.labels_
y_dbscan = dbscan.fit_predict(d)
sil_dbscan = silhouette_score(d, y_dbscan)
```

```python
plt.scatter(data1['Alcohol'], data1['Color_Intensity'], c = dbscan_labels, cmap = 'Accent', edgecolors = 'b')
plt.scatter(data1['Alcohol'], data1['Magnesium'], c = dbscan_labels, cmap = 'Accent', edgecolors = 'b')
plt.title('D-BSCAN')
plt.xlabel('Cluster Density Range')
plt.show()
```

    => Silhoutte Scores Of the Above Models:

```python
print('Score for AGGLOMERATIVE HIERARCHIAL CLUSTERING MODEL :')
print('\n''\t','The Silhouette Score with number of Clusters(3):- ', sil_avg)
print('\t','The Silhouette Score with number of Clusters(2):- ', sil_avg1)
print('\n''\n','Score for K-MEANS MODEL: ', sil_kmeans)
print('\n''\n','Score for DBSCAN MODEL: ', sil_dbscan)
```

    INFERENCE: From the above references in respect to Silhouette Scores of Models Developed, Density-Based Spatial Clustering of Applications with Noise (DBSCAN) have more score than the other two models.

# Plotting different Parameters V/s Classifications respectively;

```python

plt.figure(figsize=(12,12))
plt.title('Flora Iris V/S Petal Length', fontsize=28)
a=sns.boxenplot(x=data.Classification, y=data.PL)
a.set_xlabel('Classifications')
a.set_ylabel('Petal Length')
```

```python
plt.figure(figsize=(20,5))
plt.title('Sepal Width V/s Flora Iris', fontsize=28)
plt.scatter(x= data.SW, y=data.Classification)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.set_palette("Paired")
plt.title('Parametres V/s Flora Iris', fontsize=26)
c=sns.scatterplot(x= data.SL, y=data.Classification)
c=sns.scatterplot(x= data.SW, y=data.Classification)
c=sns.scatterplot(x= data.PW, y=data.Classification)
c=sns.scatterplot(x= data.PL, y=data.Classification)
plt.legend(data)
c.set_xlabel('Parametres')
c.set_ylabel('Classification')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.title('Flora Iris V/S Petal Width', fontsize=25)
d=sns.barplot(x=data.Classification, y=data.PW)
d.set_xlabel('Classifications')
d.set_ylabel('Petal Width');
```

```python
plt.figure(figsize=(20,5))
plt.title('Sepal Length V/S Flora Iris', fontsize=26)
e=sns.barplot(y=data.Classification, x=data.SL, orient ='h')
e.set_xlabel('Sepal Length');
```

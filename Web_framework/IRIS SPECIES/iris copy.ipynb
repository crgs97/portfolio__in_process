    --> OUTLIER MANAGEMENT.

```python
plt.figure(figsize=(28,12))
sns.boxplot(data = data.sepal_width, orient = 'h' )
plt.title('OUTLIERS IN SEPAL WIDTH ATTRIBUTE(NUMERICAL CONTINUOUS VARIABLE)', fontdict = {'family': 'Verdana', 'color':  'darkred', 'style':'italic', 'size':'24'})
plt.xlabel('Data-point Values ------------------->')
plt.grid()
```

NOTE 2: 'With Reference to Insight 2 and the above and blow results are those Outliers in Sepal Width Features'.

```python
Q1 = np.percentile(data['sepal_width'], 25, interpolation='midpoint')
Q2 = np.percentile(data['sepal_width'], 50, interpolation='midpoint')
Q3 = np.percentile(data['sepal_width'], 75, interpolation='midpoint')
print('Q1 = ',Q1)
print('Q2 = ',Q2)
print('Q3 = ',Q3)
print('Median of Sepal Width =', data.sepal_width.median())
```

```python
IQR = Q3 - Q1
print('IQR =', IQR)
low_limit = Q1 - 1.5*IQR
up_limit = Q3 + 1.5*IQR
print('The Lower Fence Value is:', low_limit)
print('The Upper Fence Value is:', up_limit)
```

```python
Outliers = []
for o in data['sepal_width']:
    if((o<low_limit) or (o>up_limit)):
        Outliers.append(o)       
print('Those Sepal Width Outliers are:', Outliers)
```

NOTE 3: 'Since Logisitc Regression Model requires  Normally Distributed Datas; Petal Length & Petal Width Features current distribution has to be changed along with Sepal Width's Outlier Treatment'.

```python
data1['petal_length'] = data1['petal_length'].apply(np.log10)
data1['sepal_width'] = data1['sepal_width'].apply(np.log10)

print('SKEWNESS OF DATA AFTER LOG TRANSFORMATIONS on Petal Length and Sepal Width: ''\n', data1.skew())

for i in data1:
    sns.displot(data = data1, x = data1[i], kde = True)
    plt.grid()
```

    --> SCALING.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data1.columns
```

NOTE 4: 'Scaling is only needed for Continuous Variables hence the needed attributes are sorted under two dataframes "no_scale" & "scale"'.

```python
no_scale = data1[['sepal_width', 'petal_length', 'species']]
scale = data1[['sepal_length', 'petal_width']]
```

```python
scaled = MinMaxScaler(feature_range = (0,1)).fit_transform(scale)
scaled_data = pd.DataFrame(scaled, columns = scale.columns)
scaled_data.head()
```

```python
Data = pd.concat([scaled_data, no_scale], axis = 1)
Data
```

                                                                                            MODEL BUILDING.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, classification_report

Data.columns
```

```python
# Spliting the Data for Training and Testing the Model 

y = Data['species']
x = Data.drop('species', axis = 1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
X_train.head()
```

> LOGISTIC REGRESSION ML MODEL.

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=165, random_state = 42)
model_lr = log_reg.fit(X_train, y_train)
y_pred_lr = model_lr.predict(X_test)

print('CLASSIFICATION REPORT:', '\n', classification_report(y_test,y_pred_lr ))
print('ACCURACY:', accuracy_score(y_test, y_pred_lr)*100,'%')
print('')
print('CONFUSION MATRIX:')
pd.crosstab(y_test, y_pred_lr, rownames=['True'], colnames=['Predicted'], margins=True)
```

> K NEAREST NEIGHBOURS ML MODEL.

```python
from sklearn.neighbors import KNeighborsClassifier
error_1= []
error_2= []
for k in range(1,20):
    knn= KNeighborsClassifier(n_neighbors=k)
    model1_knn = knn.fit(X_train,y_train)
    y_pred_knn_1= model1_knn.predict(X_train)
    error_1.append(np.mean(y_train!= y_pred_knn_1))
    y_pred_knn_2= model1_knn.predict(X_test)
    error_2.append(np.mean(y_test!= y_pred_knn_2))


plt.plot(range(1,20), error_1, 'o-', label = 'Train Data')
plt.plot(range(1,20), error_2, 'o-', label = 'Test Data')
plt.title('ERROR PLOT')
plt.xlabel('k-Value------->')
plt.ylabel('Error------->')
plt.legend()
plt.grid()
```

```python
k_matrix = []
neighbors = np.arange(1, 20)

for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    model2_knn = knn.fit(X_train, y_train)
    y_pred2_knn = model2_knn.predict(X_test)
    acc = accuracy_score(y_test,y_pred2_knn)
    k_matrix.append(acc)

plt.plot(neighbors, k_matrix,'o-')
plt.title('ACCURACY PLOT')
plt.xlabel('k-Value------->')
plt.ylabel('Accuracy------->')
plt.grid()
```

    NOTE:- From the above Error & Accuracy Plots, it could be concluded that the optimum value of k lies after 10, so "10" is taken as the n_neighbors value for model deployement.

```python
knn = KNeighborsClassifier(n_neighbors = 10)# by default metric = â€™minkowskiâ€™.  
knn_model = knn.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

print('CLASSIFICATION REPORT:', '\n', classification_report(y_test,y_pred_knn))
print('ACCURACY:', accuracy_score(y_test, y_pred_knn)*100,'%')
print('')
print('CONFUSION MATRIX:')
pd.crosstab(y_test, y_pred_knn, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
knn_1 = KNeighborsClassifier(n_neighbors = 10, metric = 'canberra')
knn_model_1 = knn_1.fit(X_train, y_train)
y_pred_knn_1 = knn_model_1.predict(X_test)

print('CLASSIFICATION REPORT:', '\n', classification_report(y_test,y_pred_knn_1))
print('ACCURACY:', accuracy_score(y_test, y_pred_knn_1)*100,'%')
print('')
print('CONFUSION MATRIX:')
pd.crosstab(y_test, y_pred_knn_1, rownames=['True'], colnames=['Predicted'], margins=True)
```

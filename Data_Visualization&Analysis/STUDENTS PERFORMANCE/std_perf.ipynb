                                                                                FEATURE ENGIEERING.

```python
dis['Aggregate Mark']=((dis['Math Score'] + dis['Reading Score'] + dis['Writing Score']))
dis.head()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = cat['Gender'], y = dis['Aggregate Mark'])
plt.xticks(rotation = 60)
plt.ylim(0, 300)
plt.axhline(y = dis['Aggregate Mark'].mean(), color = 'grey', linestyle = 'dashdot', label = "Mean Line")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = cat['Race/Ethnicity'], y = dis['Aggregate Mark'])
plt.xticks(rotation = 60)
plt.ylim(0, 300)
plt.axhline(y = dis['Aggregate Mark'].mean(), color = 'grey', linestyle = 'dashdot', label = "Mean Line")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = cat['Parental Level of Education'], y = dis['Aggregate Mark'])
plt.xticks(rotation = 60)
plt.ylim(0, 300)
plt.axhline(y = dis['Aggregate Mark'].mean(), color = 'grey', linestyle = 'dashdot', label = "Mean Line")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = cat['Lunch'], y = dis['Aggregate Mark'])
plt.xticks(rotation = 60)
plt.ylim(0, 300)
plt.axhline(y = dis['Aggregate Mark'].mean(), color = 'grey', linestyle = 'dashdot', label = "Mean Line")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x = cat['Test Preparation Course'], y = dis['Aggregate Mark'])
plt.xticks(rotation = 60)
plt.ylim(0, 300)
plt.axhline(y = dis['Aggregate Mark'].mean(), color = 'grey', linestyle = 'dashdot', label = "Mean Line")
plt.legend()
plt.show()
```

>    "INSIGHT 2:" (i) Female Students have higher Aggregate Mark than the Male Students.
                (ii) Group E (Race/ Ethnicity) Students have higher Aggregate Mark among the other Groups and Group A have the lowest Aggregate Mark.
               (iii) Students whose Parental Level of Education: "Master's Degree" have higher Aggregate Mark whereas those students whose Parental Level of Education: "High School" have the lowest Aggregate Mark.
                (iv) Students under "Standard Meals" criteria have higher Aggregate Mark than those "Free/ Reduced" Students.
                 (v) Students whoever " Completed: Test Preparation Course" have higher Aggregate Mark.

* NOTE 1: The Above Insight based only on Aggregate Score(Math + Reading + Writing).

```python
plt.figure(figsize = (18, 9))

plt.subplot(131)
plt.title('MATHS SCORE')
sns.barplot(x = cat['Test Preparation Course'], y = dis['Math Score'], hue = cat['Gender'])
plt.axhline(y = dis['Math Score'].mean(), color = 'grey', linestyle = 'dashdot', label = "Avg Maths Score")
plt.legend()
plt.grid()

plt.subplot(132)
plt.title('READING SCORE')
sns.barplot(x = cat['Test Preparation Course'], y = dis['Reading Score'], hue = cat['Gender'])
plt.axhline(y = dis['Reading Score'].mean(), color = 'grey', linestyle = 'dashdot', label = "Avg Reading Score")
plt.legend()
plt.grid()

plt.subplot(133)
plt.title('WRITING SCORE')
sns.barplot(x = cat['Test Preparation Course'], y = dis['Writing Score'], hue = cat['Gender'])
plt.axhline(y = dis['Writing Score'].mean(), color = 'grey', linestyle = 'dashdot', label = "Avg Writing Score")
plt.legend()
plt.grid()

plt.show()
```

>    "INSIGHT 3:" (i) MATHS SCORE - "Male" Students scores "more" marks in "Mathematics" than "Female" Students, even if they "have Completed" or "Not Taken" the "Test Preparation Course".
                (ii) READING SCORE - "Female" Students scores "more" marks in "Reading" than "Male" Students, even if they "have Completed" or "Not Taken" the "Test Preparation Course".
                (iv) WRITING SCORE - "Female" Students scores "more" marks in "Writing" than "Male" Students, even if they "have Completed" or "Not Taken" the "Test Preparation Course".
